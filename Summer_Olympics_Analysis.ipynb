## **Q2. Which sport is having most number of Gold Medals so far? (Top 5)**

```python
data = []

for Sport in df['Sport'].unique():
    data.append([Sport , len(df[df['Sport']  == Sport])])

pd.DataFrame(data,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False).head().plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize = (5,5))
```

## **Q4. Which player has won most number of medals? (Top 5)**

```python
df = pd.read_csv("summer.csv",engine='python')
data = []
for Medal in df['Athlete'].unique():
    data.append([Medal , len(df[df['Athlete']  == Medal])])

pd.DataFrame(data,columns = ['Athlete','Medal']).sort_values(by='Medal', ascending=False).head()
```

```python
pd.DataFrame(data,columns=['Athlete','Medal']).sort_values(by = 'Medal', ascending = False).head().plot(x = 'Athlete', y = 'Medal', kind = 'bar', figsize = (10,5),color = 'red')
```

### **Q6. In which year India won first Gold Medal in Summer Olympics?**

```python
df = pd.read_csv("summer.csv",engine = 'python')
df = df[df['Country'] == 'IND']
df = df[df['Medal'] == 'Gold']

data = []
for Year in df['Year'].unique():
    data.append([Year , len(df[df['Year']  == Year])])
pd.DataFrame(data,columns = ['Year','Medal']).sort_values(by='Year', ascending=True).head()
```

India won first Gold Medal in Summer Olympics in 1928.

### **Q8. Which sport is having most female Gold Medalists? (Top 5)**

```python
df = pd.read_csv("summer.csv",engine = 'python')
df = df[df['Medal'] == 'Gold']
df = df[df['Gender'] == 'Women']

data = []
for Sport in df['Sport'].unique():
    data.append([Sport , len(df[df['Sport']  == Sport])])
pd.DataFrame(data,columns = ['Sport','Medal']).sort_values(by='Medal', ascending= False).head()
```

```python
pd.DataFrame(data,columns=['Sport','Medal']).sort_values(by = 'Medal', ascending = False).head().plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize = (10,5),color = 'yellow')
```
